```python
#Plotting array
import matplotlib.pyplot as plt
```

```python
a=9
b=8
t=np.linspace(-np.pi,np.pi,201)
x=np.sin(a*t+np.pi/2)
y=np.sin(b*t)
```

```python
plt.plot(x,y)
plt.title('Lissajous curve')
plt.grid()
plt.show()
```

```python
#SQuares curve
t=np.linspace(-np.pi,np.pi,201)
k=np.arange(1,99)
k=2*k-1
f=np.zeros_like(t)
for i,ti in enumerate(t):
    f[i]=np.sum(np.sin(k*ti)/k)
f=(4/np.pi)*f
```

```python
plt.plot(t,f)
plt.title('Square wave')
plt.grid()
plt.show()
```

```python
#meshplot
points=np.arange(-5,5,0.01)
xs,ys=np.meshgrid(points,points)
z=np.sqrt(xs**2 + ys **2)
plt.imshow(z, cmap=plt.cm.gray)
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2+y^}$ for a grid of values")
```
